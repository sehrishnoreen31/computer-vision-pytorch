```python
import torch
import numpy as np
from torchvision import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# download data
data_destination_folder = '/kaggle/working/'
# fashion mnist data
fmnist = datasets.FashionMNIST(data_destination_folder, download=True, train=True)
```

```python
# download data
data_destination_folder = '/kaggle/working/'
# CIFAR10 dataset
cifar10 = datasets.CIFAR10(data_destination_folder, download=True, train=True)
```

```python
# train and target
# fmnist
image_train_fmnist = fmnist.data
image_target_fmnist = fmnist.targets
# cifar10
image_train_cf = cifar10.data
image_target_cf = cifar10.targets
```

```python
# inspect data
print(f'fmnist data shape: {image_train_fmnist.shape},\nunique values fmnist: {np.unique(image_target_fmnist)}\nnumber of classes in fmnist: {len(np.unique(image_target_fmnist))}\nclasses in fmnist: {fmnist.classes}\n')
```

```python
print(f'cifar10 data shape: {image_train_cf.shape},\nunique values cifar10: {np.unique(image_target_cf)}\nnumber of classes in cifar10: {len(np.unique(image_target_cf))}\nclasses in cifar10: {cifar10.classes}')
```

```python
# plot a sample from each class
# fmnist
C = len(np.unique(image_target_fmnist))  # number of classes
R = 1  # one row
fig, ax = plt.subplots(R, C, figsize=(2*C, 2))
for i, label_class in enumerate(np.unique(image_target_fmnist)):
    # get indices of all samples in class
    label_x_rows = np.where(image_target_fmnist == label_class)[0]
    # one random sample
    ix = np.random.choice(label_x_rows)
    x, y = image_train_fmnist[ix], image_target_fmnist[ix]
    # plot images
    ax[i].imshow(x, cmap='gray')
    ax[i].set_title(f"Class {y}")
    ax[i].axis("off")
plt.tight_layout()
plt.show()
```

```python
# cifar10
C = len(np.unique(image_target_cf))  # number of classes
R = 1  # one row
fig, ax = plt.subplots(R, C, figsize=(2*C, 2))
for i, label_class in enumerate(np.unique(image_target_cf)):
    # get indices of all samples in class
    label_x_rows = np.where(image_target_cf == label_class)[0]
    # one random sample
    ix = np.random.choice(label_x_rows)
    x, y = image_train_cf[ix], image_target_cf[ix]
    # plot images
    ax[i].imshow(x, cmap='gray')
    ax[i].set_title(f"Class {y}")
    ax[i].axis("off")
plt.tight_layout()
plt.show()
```
