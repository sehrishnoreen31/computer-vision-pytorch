```python
import cv2, matplotlib.pyplot as plt
%matplotlib inline
```

```python
# download image
!wget -O /kaggle/working/my_image.jpg 'https://images.unsplash.com/photo-1575936123452-b67c3203c357?q=80&w=870&auto=format&fit=crop&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D'
```

```python
# load the image 
image = cv2.imread('/kaggle/working/my_image.jpg')
# make image bgr to rgb
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image_rgb.shape)
plt.figure(figsize=(3,3))
plt.imshow(image_rgb)
plt.axis('off')
```

```python
# crop image
croped_image = image_rgb[50: 800, 50: 600]
print(croped_image.shape)
plt.figure(figsize=(3,3))
plt.imshow(croped_image)
plt.axis('off')
```

```python
# obtain specific pixels
pix = image_rgb[300:, 300:] # last 300x300 onward array
plt.imshow(pix)
plt.axis('off')
```

```python
# obtain specific pixels
pix = image_rgb[:300, :300] # first 300x300 array
plt.imshow(pix)
plt.axis('off')
```
